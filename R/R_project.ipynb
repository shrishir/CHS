Installing the required Packages

```r
install.packages("caret")
install.packages("mlbench")
install.packages("party")
```

Importing the downloaded packages

```r
library(caret)
library(mlbench)
library(party)
```

Importing the dataset to be used

```r
data(diamonds)
```

Viewing the dataset

```r
View(diamonds)
```

Applying Random Forest Classifier onto the dataset and selecting which features are to be used along with performing a 2-fold cross validation (due to time constraints in the training the classifier)

```r
set.seed(42)

model <- train(price ~ ., diamonds, method = "cforest", trControl = trainControl(method = 'cv', number = 2, verboseIter = FALSE))

model
```

Using the trained classifier to predict the price of diamonds based on the features and calculating the root mean squared error (RMSE) of the prediction

```r
prediction <- predict(model, diamonds)
error <- prediction - diamonds$price
rmse_xval <- sqrt(mean(error^2))
rmse_xval
```
